### Adding data to an index

```python
import marqo
mq = marqo.Client(url="http://0.0.0.0:8882")
```

```python
mq.get_marqo()
```

```python
if "aakash-demo" in [i.index_name for i in mq.get_indexes()['results']]:
    print("Deleting index 'aakash-demo'")
    mq.delete_index(index_name="aakash-demo")
```

```python
with open("./resume.txt", "r") as f:
    text = f.read()
text
```

```python
from nltk.tokenize import word_tokenize
words = word_tokenize(text)
```

```python
import string
for i in words:
    if i in string.punctuation:
        words.remove(i)
```

```python
for i,e in enumerate(words):
    words[i] = e.lower()
words = list(set(words))
```

```python
len(words)
```

```python
documents_to_insert = []
for word in words:
    documents_to_insert.append(
        {
            "word": word
        }
    )
```

```python
mq.create_index("aakash-demo")
mq.index("aakash-demo").add_documents(
    documents=documents_to_insert,
    tensor_fields=["word"]
)
```

```python
results = mq.index("aakash-demo").search(q="Artificial Intelligence", limit=300)
```
